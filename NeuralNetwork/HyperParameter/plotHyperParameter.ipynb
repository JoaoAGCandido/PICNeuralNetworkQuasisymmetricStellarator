```python
import joblib
import matplotlib.pyplot as plt
import pandas as pd

batch_size = joblib.load("nfp3/batch_size.pkl")
print("best parameter: ", batch_size.best_params_)
df = pd.DataFrame(batch_size.cv_results_)
plt.scatter(df["param_batch_size"] , df["mean_test_score"])
plt.errorbar(df["param_batch_size"] , df["mean_test_score"], yerr = df["std_test_score"],fmt='o')
plt.xlabel('Batch size')
plt.ylabel('Mean test score')
```

```python
import joblib
import matplotlib.pyplot as plt
import pandas as pd

hiddenLayer = joblib.load("nfp3/hiddenLayer.pkl")
print("best parameter: ", hiddenLayer.best_params_)
df = pd.DataFrame(hiddenLayer.cv_results_)
df["param_hidden_layer_sizes"].astype(str)
df=df.sort_values("mean_test_score", ascending=False)
print(df.loc[:,["param_hidden_layer_sizes","mean_test_score", "std_test_score", "mean_fit_time"]])

```
