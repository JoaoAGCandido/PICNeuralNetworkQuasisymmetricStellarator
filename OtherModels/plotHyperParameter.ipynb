```python
import joblib
import matplotlib.pyplot as plt
import pandas as pd

degree = joblib.load("randSearch.pkl")
print("best parameter: ", degree.best_params_)
df = pd.DataFrame(degree.cv_results_)
#print(df)
plt.scatter(df["param_poly__degree"].iloc[0:16] , df["mean_test_score"].iloc[0:16])
plt.errorbar(df["param_poly__degree"].iloc[0:16] , df["mean_test_score"].iloc[0:16], yerr = df["std_test_score"].iloc[0:16],fmt='o')
plt.xlabel('Degree')
plt.ylabel('Mean test score')
df.sort_values("param_poly__degree", inplace=True)
print(df.loc[:,["param_poly__degree","mean_test_score","mean_fit_time"]])
```
